```python
import numpy as np
import pandas as pd
```

```python
registry = pd.read_excel("datasets/Registry.xlsx", nrows=20000)
```

```python
registry.head(5)
```

```python
# удаляем 1-ый столбец:
registry.drop(columns="Единый реестр субъектов малого и среднего предпринимательства по состоянию на 15.04.2024", inplace=True)
# удаляем 1-ую строку:
registry.drop(0, inplace=True)
```

```python
registry.head(3)
```

```python
# именуем столбцы
dic = {}
old = registry.keys()
new = registry.iloc[0].values
for i in range(len(old)): dic[old[i]] = new[i]
registry.rename(columns=dic, inplace=True)

#удаляем исходную строку с заголовками столбцов
registry = registry.drop(1).reset_index().drop(columns="index")
```

Удаляем ненужные колонки и колонки, состоящие только из одного значения (в таблице выше -- строка `unique`):

```python
registry.drop(
    columns=[
        "Производство инновационной, высокотехнологичной продукции",
        "Дата исключения из реестра", "Вновь созданный",
        "Наличие заключенных договоров, контрактов",
        "Участие в программах партнерства", "ОГРН",
        "Район", "Телефон", "E-mail", "WWW",
        "Является социальным предприятием"
    ], inplace=True)
```

```python
registry.sample(3)
```

```python
registry.describe()
```

```python
mask = (registry["Тип субъекта"] == "Юридическое лицо") \
    & (registry["Основной вид деятельности"] == "41.20 Строительство жилых и нежилых зданий") & (
        (registry["Категория"] == "Малое предприятие") | (registry["Категория"] == "Среднее предприятие")
)
registry = registry[mask].reset_index().drop(columns="index")
registry
```

```python
registry.drop(columns=["Тип субъекта", "Основной вид деятельности"], inplace=True)
registry.to_csv("datasets/ClearedRegistry.csv")
```
